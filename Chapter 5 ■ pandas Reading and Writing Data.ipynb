Using RegExp for Parsing TXT Files


```python
pd.read_table('ch05_04.txt',sep=' ')
```

Remember to use the header option set to None whenever the column headings are not present in the 
TXT file.

```python
pd.read_table('ch05_05.txt',sep='\D*',header=None, engine='python')
```

```python
pd.read_table('ch05_06.txt',sep=',',skiprows=[0,1,3,6]) # skiprows option you can exclude all the lines you want
```
